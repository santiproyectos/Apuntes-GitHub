 Instrucciones

Para cada elemento de la lista:

Salida esperada

{1: 2, 2: 3, 3: 1, 4: 2, 5: 1}


```python
lista = [1,2,3,4,3,12,5,2,1,5,6]
def contar_frecuencia(lista):
    frecuencia = {}
    i = 0
    while i < len(lista):  
        numero = lista [i]  
        if numero in frecuencia:
            frecuencia[numero] += 1  
        else:
            frecuencia[numero] = 1  
        i += 1
    return frecuencia   
resultado = contar_frecuencia(lista)
print(resultado) 
```
