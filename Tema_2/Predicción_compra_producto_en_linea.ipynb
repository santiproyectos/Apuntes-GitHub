Realiza predicciones sobre nuevos usuarios usando predecir_compra(modelo, 8, 12) para predecir si un usuario que vio 8 p치ginas y pas칩 12 minutos en el sitio har치 una compra.





```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Paso 1: Generaci칩n de Datos Sint칠ticos de Compras
def generar_datos_compras(num_muestras):
    """
    Genera un conjunto de datos sint칠ticos con el comportamiento de los usuarios en un sitio web.
    Cada usuario tiene dos caracter칤sticas: n칰mero de p치ginas vistas y tiempo en el sitio.
    La etiqueta indica si el usuario comprar치 o no el producto (1 o 0).
    """
    X = []  # Caracter칤sticas: num_paginas_vistas, tiempo_en_sitio
    y = []  # Etiqueta: 1 (compra), 0 (no compra)
    
    for _ in range(num_muestras):
        num_paginas_vistas = np.random.randint(1, 21)  # N칰mero de p치ginas entre 1 y 20
        tiempo_en_sitio = np.random.uniform(0, 30)     # Tiempo entre 0 y 30 minutos
        
        # Etiqueta: 1 si num_paginas_vistas > 5 y tiempo_en_sitio > 10, 0 en caso contrario
        if num_paginas_vistas > 5 and tiempo_en_sitio > 10:
            compra = 1
        else:
            compra = 0
        
        X.append([num_paginas_vistas, tiempo_en_sitio])
        y.append(compra)
    
    return np.array(X), np.array(y)

# Paso 2: Entrenamiento del Modelo
def entrenar_modelo(X, y):
    """
    Entrena un modelo de Machine Learning para predecir la compra de un producto.
    Utiliza un modelo de regresi칩n log칤stica para clasificaci칩n.
    """
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    
    modelo = LogisticRegression()
    modelo.fit(X_train, y_train)
    
    # Evaluar el modelo en el conjunto de prueba
    y_pred = modelo.predict(X_test)
    precision = accuracy_score(y_test, y_pred)
    
    print(f"Precisi칩n del modelo: {precision:.2f}")
    
    # Devolvemos solo el modelo entrenado
    return modelo

# Paso 3: Predicci칩n de Compra
def predecir_compra(modelo, num_paginas_vistas, tiempo_en_sitio):
    """
    Predice si un usuario comprar치 el producto basado en el n칰mero de p치ginas vistas y el tiempo en el sitio.
    """
    prediccion = modelo.predict([[num_paginas_vistas, tiempo_en_sitio]])
    
    # Convertir la predicci칩n en un mensaje textual
    if prediccion[0] == 1:
        return 'El usuario comprar치 el producto.'
    else:
        return 'El usuario no comprar치 el producto.'

# Ejemplo de Ejecuci칩n:

# 1. Generar datos sint칠ticos de usuarios
X, y = generar_datos_compras(100)

# 2. Entrenar el modelo
modelo = entrenar_modelo(X, y)

# 3. Realizar una predicci칩n para un nuevo usuario
num_paginas_vistas = 8  # El usuario vio 8 p치ginas
tiempo_en_sitio = 12  # El usuario estuvo en el sitio por 12 minutos
prediccion = predecir_compra(modelo, num_paginas_vistas, tiempo_en_sitio)

# Imprimir el mensaje final de la predicci칩n
print(prediccion)
```
