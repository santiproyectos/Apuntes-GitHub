```python
import numpy as np
def generate_dataset(n=100):
    np.random.seed(42)
    X = []
    y = []
    for _ in range(n):
        velocidad = np.random.rand()
        mantenimiento = np.random.rand()
        bibliotecas = np.random.rand()
        tipo_app = np.random.randint(0, 3)  
        rendimiento = np.random.rand()
        features = [velocidad, mantenimiento, bibliotecas, tipo_app, rendimiento]
        if rendimiento > 0.8 and tipo_app == 2:
            etiqueta = 3  # C++
        elif mantenimiento > 0.7 and tipo_app == 0:
            etiqueta = 0  # Python
        elif velocidad > 0.6 and tipo_app == 1:
            etiqueta = 2  # Java
        else:
            etiqueta = 1  # JavaScript
        X.append(features)
        y.append(etiqueta)
    return np.array(X), np.array(y)
from sklearn.ensemble import RandomForestClassifier
class LanguagePredictor:
    def __init__(self):
        self.modelo = RandomForestClassifier(n_estimators=100, random_state=42)
        self.label_map = {0: "Python",1: "JavaScript",2: "Java",3: "C++"}
    def train(self, X, y):
        self.modelo.fit(X, y)
    def predict(self, features: np.ndarray) -> str:
        pred = self.modelo.predict([features])[0]
        return self.label_map[pred]
X, y = generate_dataset()
predictor = LanguagePredictor()
predictor.train(X, y)
new_project = np.array([0.7, 0.9, 0.5, 1, 0.6])  
pred = predictor.predict(new_project)
print(f"Lenguaje recomendado para el nuevo proyecto: {pred}")
```
