```python
class PlayerMatchData:
    def __init__(self, kills, deaths, assists, damage_dealt, damage_received, healing_done, objective_time, won):
        self.kills = kills
        self.deaths = deaths
        self.assists = assists
        self.damage_dealt = damage_dealt
        self.damage_received = damage_received
        self.healing_done = healing_done
        self.objective_time = objective_time
        self.won = won  
    def to_dict(self, include_target=True):
        data = {'kills': self.kills,'deaths': self.deaths,'assists': self.assists,'damage_dealt': self.damage_dealt,'damage_received': self.damage_received,'healing_done': self.healing_done,'objective_time': self.objective_time}
        if include_target and self.won is not None:
            data['won'] = self.won
        return data
import numpy as np
def generate_synthetic_data(n=100):
    data = []
    for _ in range(n):
        kills = np.random.poisson(5)
        deaths = np.random.poisson(3)
        assists = np.random.poisson(2)
        damage_dealt = kills * 300 + np.random.normal(0, 100)
        damage_received = deaths * 400 + np.random.normal(0, 100)
        healing_done = np.random.randint(0, 301)
        objective_time = np.random.randint(0, 121)
        won = int((damage_dealt > damage_received) and (kills > deaths))
        player = PlayerMatchData(kills, deaths, assists, damage_dealt, damage_received, healing_done, objective_time, won)
        data.append(player)
    return data
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
import pandas as pd
class VictoryPredictor:
    def __init__(self):
        self.model = LogisticRegression()
        self.scaler = StandardScaler()
    def train(self, data):
        df = pd.DataFrame([player.to_dict() for player in data])
        X = df.drop('won', axis=1)
        y = df['won']
        X_scaled = self.scaler.fit_transform(X)
        self.model.fit(X_scaled, y)
    def predict(self, player: PlayerMatchData):
        player_df = pd.DataFrame([player.to_dict(include_target=False)])
        player_scaled = self.scaler.transform(player_df)
        return self.model.predict(player_scaled)[0]
training_data = generate_synthetic_data(150)
predictor = VictoryPredictor()
predictor.train(training_data)
test_player = PlayerMatchData(8, 2, 3, 2400, 800, 120, 90, None)
prediction = predictor.predict(test_player)
print(f"쮼l jugador ganar치? {'S칤' if prediction == 1 else 'No'}")
        
```
