```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier

# Paso 1: Generaci칩n de Datos de Frutas
def generar_datos_frutas(num_muestras):
    """
    Genera un conjunto de datos ficticio con 'num_muestras' frutas.
    Cada fruta tiene un peso (en gramos) y un tama침o (en cm).
    Asigna una etiqueta seg칰n el tipo de fruta.
    """
    # Caracter칤sticas: peso (en gramos) y tama침o (en cm)
    # Etiquetas: 'Manzana', 'Pl치tano', 'Naranja'
    frutas = []
    etiquetas = []
    
    for _ in range(num_muestras):
        peso = np.random.uniform(100, 300)  # Peso aleatorio entre 100 y 300 gramos
        tama침o = np.random.uniform(5, 20)   # Tama침o aleatorio entre 5 y 20 cm
        
        # Etiqueta seg칰n el peso y tama침o
        if peso < 150 and tama침o < 10:
            fruta = 'Manzana'
        elif peso > 150 and tama침o > 10:
            fruta = 'Naranja'
        else:
            fruta = 'Pl치tano'
        
        frutas.append([peso, tama침o])
        etiquetas.append(fruta)
    
    return np.array(frutas), np.array(etiquetas)

# Paso 2: Entrenamiento del Modelo
def entrenar_modelo(data):
    """
    Entrena un modelo de Machine Learning para clasificar frutas seg칰n su peso y tama침o.
    Usamos un RandomForestClassifier para este prop칩sito.
    """
    # Generar los datos
    X, y = data
    
    # Inicializar y entrenar el modelo
    modelo = RandomForestClassifier(n_estimators=100, random_state=42)
    modelo.fit(X, y)
    
    return modelo

# Paso 3: Predicci칩n de una Nueva Fruta
def predecir_fruta(modelo, peso, tama침o):
    """
    Usa el modelo entrenado para predecir a qu칠 tipo de fruta pertenece un nuevo ejemplo.
    """
    # Hacer una predicci칩n
    prediccion = modelo.predict([[peso, tama침o]])
    
    return prediccion[0]

# Ejemplo de ejecuci칩n:

# Generar un conjunto de datos ficticio con 1000 muestras
data = generar_datos_frutas(1000)

# Entrenar el modelo con los datos generados
modelo = entrenar_modelo(data)

# Predecir la fruta de una nueva muestra
peso_nueva_fruta = 170  # peso en gramos
tama침o_nueva_fruta = 15  # tama침o en cm

fruta_predicha = predecir_fruta(modelo, peso_nueva_fruta, tama침o_nueva_fruta)

# Mostrar la predicci칩n
print(f"La fruta con peso {peso_nueva_fruta} g y tama침o {tama침o_nueva_fruta} cm es un(a) {fruta_predicha}.")
```
